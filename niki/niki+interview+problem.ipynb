```python
words=[]
pos=[]
for i in data['question']:
    text=nltk.word_tokenize(i)
    words.append(text[0])
    pos.append(nltk.pos_tag(text))
```

```python
l2=list(set(words))
```

```python
def index1(word):
    if word in l2:
        return l2.index(word)
    else:
        return 999
 
```

```python
def index2(pos2):
    if pos2 in pos:
        return pos.index(pos2)
    else:
        return 999
        
```

```python
from sklearn.cross_validation import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(data['question'],data['label'],test_size=0.3,random_state=123)
```

```python
# Going by Emperical way instead of any embeddings .. No Named Entity tagging done even..Only Used pos of words at certain position 
#of the sentence for initial run
def transform(l3):
    last_word_pos=[]
    first_word_pos=[]
    mid_word_pos=[]
    first_word=[]
    sec_last_word_pos=[]
    sec_word_pos=[]
    for i in l3:
        i = i.lower()
        word=nltk.word_tokenize(i) 
        word2=nltk.pos_tag(word)
        #print word2
        l=len(word)
        last_word_pos.append(index2(word2[l-2]))
        sec_last_word_pos.append(index2(word2[l-3]))
        first_word_pos.append(index2(word2[0]))
        sec_word_pos.append(index2(word2[1]))
        mid_word_pos.append(index2(word2[(l-1)/2]))
        first_word.append(index1(word[0]))
    new_data=pd.DataFrame({'last_word_pos':last_word_pos,'sec_last_word_pos':sec_last_word_pos,'sec_word_pos':sec_word_pos,'first_word_pos':first_word_pos,'mid_word_pos':mid_word_pos,'first_word':first_word})
    #new_data['first_word']=new_data['first_word'].astype('category')
    #return new_data.iloc[:,:1]
    return new_data
```

```python
#x_train,x_test,y_train,y_test=train_test_split(new_data,data['label'],test_size=0.2,random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier( n_estimators = 101,verbose=1 )
y_pred = forest.fit(transform(list(data['question'])),data['label'])
```

```python
# Alternative fast way
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(transform(list(data['question'])),data['label'])
```
